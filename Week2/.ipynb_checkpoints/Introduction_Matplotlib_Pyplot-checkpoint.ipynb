<h3> Customizing Your Plots </h3>

<ul>
    <li><b>legend()</b>To add a legend</li>
    <li><b>axis()</b> Adjust axes</li>
    <li><b>xlabel(), ylabel()</b>Set axis labels</li>
    <li><b>savefig()</b>To save figure</li>
</ul>

```python
x = np.linspace(0, 10, 20)
y1 = x**2.0
y2 = x**1.5
plt.plot(x, y1, "bo-", linewidth=2, markersize=5, label="First")
plt.plot(x, y2, "gs-", linewidth=2, markersize=5, label="Second")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
plt.savefig("myplot.jpeg")   ## Saving our plot
```

<h3> Plotting Using Logarithmic Axes  </h3>

```python
## Using loglog
x = np.linspace(0, 10, 20)
y1 = x**2.0
y2 = x**1.5
plt.loglog(x, y1, "bo-", linewidth=2, markersize=5, label="First")
plt.loglog(x, y2, "gs-", linewidth=2, markersize=5, label="Second")
plt.xlabel("X")
plt.ylabel("Y")
#plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
```

```python
## Using loglog
x = np.logspace(-1, 1, 40)
y1 = x**2.0
y2 = x**1.5
plt.loglog(x, y1, "bo-", linewidth=2, markersize=5, label="First")
plt.loglog(x, y2, "gs-", linewidth=2, markersize=5, label="Second")
plt.xlabel("X")
plt.ylabel("Y")
#plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
```

```python
x = np.logspace(0,1,10)
y = x**2
plt.loglog(x,y,"bo-")
```

<h3> Generating Histograms </h3>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.normal(size=1000)
```

```python
plt.hist(x)
```

```python
plt.hist(x, normed=True);
```

```python
plt.hist(x, bins=np.linspace(-5, 5, 21));
```

```python
x = np.random.gamma(2, 3, 100000)
```

```python
plt.figure()
plt.subplot(221)
plt.hist(x, bins = 30);
plt.subplot(222)
plt.hist(x, bins = 30);
plt.subplot(223)
plt.hist(x, bins = 30, cumulative = 30);
plt.subplot(224)
plt.hist(x, bins = 30, cumulative = 30, histtype = "step");
```
